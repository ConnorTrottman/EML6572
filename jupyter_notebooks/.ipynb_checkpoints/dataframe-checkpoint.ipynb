```python
import csv
from datetime import datetime
from pprint import pprint # pretty printing
```

```python
## Define function print first point
def print_first_point(filename):
    """
    This function prints and returns the first data point (second row) from
    a csv file that includes a header row.
    """
    # print dataset name for reference
    dataset = filename.split('.csv')[0].split('/')[-1]
    print('\ndataset: {}'.format(dataset))
    
    with open(filename, 'r') as f_in:
        ## Use the csv library to set up a DictReader object. ##
        ## see https://docs.python.org/3/library/csv.html           ##
        trip_reader = csv.DictReader(f_in)
        
        ## Use a function on the DictReader object to read the     ##
        ## first trip from the data file and store it in a variable.     ##
        ## see https://docs.python.org/3/library/csv.html#reader-objects ##
        first_trip = trip_reader.__next__()
        
        ## Use the pprint library to print the first trip. ##
        ## see https://docs.python.org/3/library/pprint.html     ##
        pprint(first_trip)
        
    # output dataset name and first trip for later testing
    return (dataset, first_trip)

# list of files
data_files = ['../data/tripdata/Q1_2018.csv', '../data/tripdata/Q2_2018.csv',
             '../data/tripdata/Q1_2017.csv','../data/tripdata/Q2_2017.csv',
             '../data/tripdata/Q3_2017.csv','../data/tripdata/Q4_2017.csv',
             '../data/tripdata/Q3_2016.csv','../data/tripdata/Q4_2016.csv',
             '../data/stationdata/stations.csv', 
             '../data/financialdata/financial_cleanup.csv']

# print the first trip from each file, store in dictionary
example_trips = {}
for data_file in data_files:
    dataset, first_trip = print_first_point(data_file)
    example_trips[dataset] = first_trip
```
